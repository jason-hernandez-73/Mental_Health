```python
#predict=mental_illness
#we try to the outstanding variable that has the highest correlation to our prediction 
selected_df=clean_df[["employed","mental_illness","Education","hospitalized","disabled","with_parents","gap","months_gap","income_with_social","read","welfare_income","food_stamp","section_8","Lack_concentration","Anxiety","Depression","Obsessive_thinking","Mood_swings","Panic_attacks","Compulsive_behavior","Tiredness","Age","Gender","Household_Income"]]
selected_df
s_df=selected_df.fillna(0)
s5_df = s_df[s_df.Household_Income != "Prefer not to answer"]
s5_df
```

```python
def locedu(df):
    df=df.assign(educ_label='0')
    df.loc[df['Education']=='High School or GED','educ_label']='1'
    df.loc[df['Education']=='Some Undergraduate','educ_label']='2'
    df.loc[df['Education']=='Completed Undergraduate','educ_label']='3'
    df.loc[df['Education']=='Completed Masters','educ_label']='4'
    df.loc[df['Education']=='Some Phd','educ_label']='5'
    df.loc[df['Education']=='Complete Phd','educ_label']='6'
    return df
s6_df=locedu(s5_df)
def locincome(df):
    df=df.assign(income_label='0')
    df.loc[df['Household_Income']=='$10,000-$24,999','income_label']='1'
    df.loc[df['Household_Income']=='$25,000-$49,999','income_label']='2'
    df.loc[df['Household_Income']=='$50,000-$74,999','income_label']='3'
    df.loc[df['Household_Income']=='$75,000-$99,999','income_label']='4'
    df.loc[df['Household_Income']=='$100,000-$124,999','income_label']='5'
    df.loc[df['Household_Income']=='$125,000-$149,999','income_label']='6'
    df.loc[df['Household_Income']=='$150,000-$174,999','income_label']='7'
    df.loc[df['Household_Income']=='$175,000-$199,999','income_label']='8'
    df.loc[df['Household_Income']=='$200,000+','income_label']='9'
    return df
s7_df=locincome(s6_df) 
def locgender(df):
    df=df.assign(gender_label='0')
    df.loc[df['Gender']=='Male','gender_label']='1'
    return df
s8_df=locgender(s7_df)
def locage(df):
    df=df.assign(age_label='0')
    df.loc[df['Age']=='30-44','age_label']='1'
    df.loc[df['Age']=='45-60','age_label']='2'
    df.loc[df['Age']=='> 60','age_label']='3'
    return df
s9_df=locage(s8_df)
s10_df=s9_df[["employed","mental_illness","hospitalized","disabled","with_parents","gap","months_gap","income_with_social","read","welfare_income","food_stamp","section_8","Lack_concentration","Anxiety","Depression","Obsessive_thinking","Mood_swings","Panic_attacks","Compulsive_behavior","Tiredness","educ_label","income_label","age_label","gender_label"]]

s10_df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection, naive_bayes, metrics
def train_model(classifier, feature_vector_train, train_label, feature_vector_valid, valid_label, is_neural_net=False):
    classifier.fit(feature_vector_train, train_label)
    predictions=classifier.predict(feature_vector_valid)
    if is_neural_net:
        predictions=predictions.argmax(axis=-1)
    return metrics.accuracy_score(predictions, valid_label)

```
